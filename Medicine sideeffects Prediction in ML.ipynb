```python
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
```

```python
train_labels = []
train_samples = []
```

```python
# Generating Data 
# creating our own data
for i in range(50):
    # 5% younger individuals experiencing sideeffects
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(1)      # 1 means they are affected from medicine
   
    # 5 % of older didnot experiencing side effects
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(0)      # 0 means not affect by sideeffect

for i in range(1000):
    # 95 % younger individuals not experiencing side effects
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(0)
    
    #95% older experiencing side effect
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(1)
```

```python
for i in train_samples:
    print(i)
```

```python
for i in train_labels:
    print(i)
```

```python
# converting both those train samples into numpy arrays 
train_labels = np.array(train_labels)
train_samples = np.array(train_samples)
```

```python
train_labels.size
```

```python
train_samples.size
```

```python
# shuffling to get rid of orders in the dataset during creation process
train_labels, train_samples = shuffle(train_labels,train_samples)
```

```python
# feature scaling from age 13 to 100 to 0 to 1 because it helps in ML
scaler = MinMaxScaler(feature_range=(0,1))
scaled_train_samples = scaler.fit_transform(train_samples.reshape(-1,1))
```

```python
#print scale data
for i in scaled_train_samples:
    print(i)
```

```python
import tensorflow as tf
from tensorflow import keras 
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Activation, Dense 
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
```

```python
# building a model 

model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=16, input_shape=(1,),activation='relu'),
    tf.keras.layers.Dense(units=32, activation='relu'),
    tf.keras.layers.Dense(units=2, activation='softmax')
    
])
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.0001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# training the model 
# data runs through validation statte where we'll know it is not overfitting and has high bias or variance 
model.fit(x=scaled_train_samples, y=train_labels,validation_split=0.1, batch_size=10,epochs=30, shuffle=True, verbose=2)
```

```python
# Creating a test data like train data 
test_labels = []
test_samples = []
```

```python
for i in range(10):
    # 5% younger individuals experiencing sideeffects
    random_younger = randint(13,64)
    test_samples.append(random_younger)
    test_labels.append(1)      # 1 means they are affected from medicine
   
    # 5 % of older didnot experiencing side effects
    random_older = randint(65,100)
    test_samples.append(random_older)
    test_labels.append(0)      # 0 means not affect by sideeffect

for i in range(200):
    # 95 % younger individuals not experiencing side effects
    random_younger = randint(13,64)
    test_samples.append(random_younger)
    test_labels.append(0)
    
    #95% older experiencing side effect
    random_older = randint(65,100)
    test_samples.append(random_older)
    test_labels.append(1)
```

```python
test_labels = np.array(test_labels)
test_samples = np.array(test_samples)
test_labels, test_samples = shuffle(test_labels,test_samples)
```

```python
scaled_test_samples = scaler.fit_transform(test_samples.reshape(-1,1))
```

```python
# PREDICTING THE MODEL 
predictions = model.predict(x=scaled_test_samples,batch_size=10,verbose=2)
```

```python
for i in predictions:
    print(i)
```

```python
# rounding off the predictions in 0 and 1
rounded_predictions = np.argmax(predictions, axis=-1)
```

```python
for i in rounded_predictions:
    print(i)
```

```python
%matplotlib inline 
from sklearn.metrics import confusion_matrix
import itertools
import matplotlib.pyplot as plt
```

```python
# the value of true value goes from the testlabel and predicted value goes from rounded prediction
cm = confusion_matrix(y_true = test_labels, y_pred = rounded_predictions)
```

```python
# This function prints and plots confusion matrix 
# And we will normalize the value by using normalization
# this code is directly extracted from scikit code webside for plotting confusion matrix on graph

def plot_confusion_matrix(cm, classes,
                        normalize=False,
                        title='Confusion matrix',
                        cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
            horizontalalignment="center",
            color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
cm_plot_labels = ['no side effects', 'had side effects']
plot_confusion_matrix(cm=cm,classes=cm_plot_labels,title='Confusion Matrix')
```
